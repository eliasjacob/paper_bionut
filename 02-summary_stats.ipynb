```python
# Summary stats requested by Dr. Michelle Jacob on 2024-10-20

import pandas as pd
# Change pandas to show only 2 decimal places
pd.options.display.float_format = '{:.2f}'.format

# Change pandas to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('input/Supplementary Material 3 - BIONUT - Neglected Food Species.csv')
```

```python
df
```

```python
df['total number of recipes'] = df['number of recipes (POF/IBGE)'] + df['number of recipes - other sources']
df.query('`total number of recipes`.isnull()', engine='python')
```

```python
df['total number of recipes'] = df['total number of recipes'].fillna(0)
df['total number of recipes'] = df['total number of recipes'].astype(int)
```

```python
df.columns
```

```python


request_1 = ['food category', 'conservation status', 'origin', 'commercial cultivation', 'availability of nutritional composition data', 'total number of recipes']

for col in request_1:
    print(df[col].value_counts(normalize=True, dropna=False) * 100)
    print('\n')
```

```python
df['total number of recipes'].describe([0.1, 0.2, 0.25, 0.5, 0.75, 0.9])
```

```python
request_2 = ['conservation status', 'origin', 'commercial cultivation', 'availability of nutritional composition data', 'total number of recipes']
for col in request_2:
    print(df.groupby('food category')[col].value_counts(normalize=True, dropna=False) * 100)
    print('\n')
```

```python
df.groupby('food category')['total number of recipes'].describe([0.1, 0.2, 0.25, 0.5, 0.75, 0.9])
```
